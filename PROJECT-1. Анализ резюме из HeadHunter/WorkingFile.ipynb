```python

def experience(data):
    if type(data) == float:
        return np.nan
    else:
        
        sp_data = data.split(' ')
        sp_data = sp_data[:6]

        # Создадим вспомогательные списки с возможным отображением лет и месяцев
        year_list = ['год', 'года', 'лет']
        month_list = ['месяц', 'месяца', 'месяцев']

        res_year = 0
        res_month = 0

        for i in range(len(sp_data)):
            if sp_data[i] in year_list:
                res_year = sp_data[i-1]
            if sp_data[i] in month_list:
                res_month = sp_data[i-1]
            
        result = int(res_year) * 12 + int(res_month)

        if result == 0:
            return np.nan
        else:
            return result

data_hh['Опыт работы (месяц)'] = data_hh['Опыт работы'].apply(experience)
data_hh['Опыт работы (месяц)'].median()

    
```
